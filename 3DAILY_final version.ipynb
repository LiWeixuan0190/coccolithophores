## Number of times of rank 1

### BATS points

```julia
# establish tracers name array
la=[31.5,32.5,33.5,35.5,37.5,38.5,39.5]
lat = ncread(path3d[1],"Y")
nm=[] 
for i in 1:39
        a = 60+i
        nm=push!(nm,string(a))
end
    push!(nm,"0a")
    push!(nm,"0b")
    push!(nm,"0c")
    push!(nm,"0d")
    push!(nm,"0e")
    push!(nm,"0f")
    push!(nm,"0g")
    push!(nm,"0h")
    push!(nm,"0i")
    push!(nm,"0j")
    push!(nm,"0k")

for n in 1:7
index_lat = findall(x->x==la[n], lat)
tm=zeros(120)
p_total=zeros(5,10,120)
p_copy=zeros(5,10)
row=zeros(120)
col=zeros(120)
for i in 1:120
    for j in 1:5
    for k in 1:10
        t = ncread(path3d[i],"T")
        tm[i] = Int(t[1]/3600.0/24.0-1442.0)
        nmt=nm[10*(j-1)+k]
        p_temp = ncread(path3d[i],"TRAC"*nmt)
        p_total[j,k,i]=sum(p_temp[289:296,index_lat[1],1,1])
    end
    end
    p_copy=p_total[:,:,i]
    maxval1, maxind1=findmax(p_copy)
    row[i],col[i]=Tuple(maxind1) # row is cell size, col is thermal type
end
rank1=zeros(5,10)
    for i in 1:120
        rank1[Int(row[i]),Int(col[i])]=rank1[Int(row[i]),Int(col[i])]+1
    end 
p_plot=heatmap(rank1,aspect_ratio=:equal,
color=:viridis,title="Number of times of rank first of different types \n of coccolithophores at "*string(la[n])*"N", titlefontsize=12,
xlabel="Thermal type", xlabelfontsize=10,
ylabel="Cell size", ylabelfontsize=10, ylim=(0.5,5.5))

savefig(p_plot,"/home/weixuan/Desktop/Jupyter notebook/biomass results/Times of rank1/Number of times of rank first of different types of coccolithophores at "*string(la[n])*"N.png")
end
```

## Coexistence times of each location and each time

### Final code

```julia
file_path = "/home/akrinos/coccolithophores/no_dask_process/"
coexm = zeros(50,50)
for num in 1:20
    num1 = (num-1)*5
    num2 = num1+5
    test_path = joinpath(file_path,"coccolithophore_data_output_step_w_"*string(num1)*"_"*string(num2)*".csv")

    bio = CSV.File(test_path,select=[2,3,4,8,9])
    bio_matrix = bio |> Tables.matrix
    bio_sort=bio_matrix[reverse(sortperm(bio_matrix[:,1])),:]
    bio_de=bio_sort[bio_sort[:,1].>0.01,:]
    bio_final=bio_de[sortperm(bio_de[:,2]),:]
    bio_loc=zeros(size(bio_final,1),6)
    bio_loc[:,1:5]=bio_final
    bio_loc[:,6]=bio_final[:,2]*10000000+bio_final[:,4]*10000+bio_final[:,5]
    s=bio_loc[sortperm(bio_loc[:,6]),:]
    for i in 1:size(s,1)
        s[i,3]=s[i,3]+1
    end
    
    for p in 1:size(s,1)
        coexm[Int.(s[p,3]),Int.(s[p,3])] = coexm[Int.(s[p,3]),Int.(s[p,3])]+1
    end
    

    temp = zeros(50)
    n=1
    temp[n]=s[1,3]
    for i in 2:size(s,1)
        if s[i,6]==s[i-1,6]
            n=n+1;
            temp[n]=s[i,3]
            for x = 1:n-1
                y = n
                coexm[Int.(temp[x]),Int.(temp[y])]=coexm[Int.(temp[x]),Int.(temp[y])]+1
            end
        end

        if s[i,6]!=s[i-1,6]
           n=1
           temp[n]=s[i,3]
        end
    end
   
    if s[size(s,1),6] != s[size(s,1)-1,6]
        coexm[Int.(s[size(s,1),3]),Int.(s[size(s,1)-1,3])]=coexm[Int.(s[size(s,1),3]),Int.(s[size(s,1)-1,3])]+1
    end
                            
    for i in 1:49
        for j in i+1:50
            coexm[i,j]=coexm[i,j]+coexm[j,i];
            coexm[j,i]=0.0;
        end
    end
    end
```
