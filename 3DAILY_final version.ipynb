### Biomass over the whole ocean

```julia
lat = ncread(path3d[1],"Y")
lon = ncread(path3d[1],"X")
nm=[]
# establish tracers name array
    for i in 1:39
        a = 60+i
        nm=push!(nm,string(a))
    end
    push!(nm,"0a")
    push!(nm,"0b")
    push!(nm,"0c")
    push!(nm,"0d")
    push!(nm,"0e")
    push!(nm,"0f")
    push!(nm,"0g")
    push!(nm,"0h")
    push!(nm,"0i")
    push!(nm,"0j")
    push!(nm,"0k")

tm=zeros(120)
p1=zeros(120,10)
p_total=zeros(5,10,120)
for i in 1:120
for j in 1:5
for k in 1:10
    t = ncread(path3d[i],"T")
    tm[i] = t[1]/3600.0/24.0-1442.0
    nmt=nm[10*(j-1)+k]
    p_temp = ncread(path3d[i],"TRAC"*nmt)
    p_total[j,k,i]=sum(p_temp[:,:,1,1])
end
end
end
p1=sum(p_total,dims=1)
p2=zeros(10,120)
p2=transpose(p1[1,:,:])
pplot=plot(tm,p2,title="Biomass variation across a year of coccolithophores with different thermal types \n over the whole ocean",
titlefontsize = 9,
label = "",
color = [:firebrick :red1 :orange :yellow1 :lime :green :darkslategray1 :skyblue :royalblue :blue2],
lw=2,xlabel="Days",ylabel="Concentration (kg/kg)")
    
savefig(pplot,"/home/weixuan/Desktop/Jupyter notebook/biomass results/Biomass global variation/Biomass variation across a year of coccolithophores with different thermal types \n over the whole ocean.png")
```

### Biomass at BATS points

```julia
la=[31.5,32.5,33.5,35.5,37.5,38.5,39.5]
lat = ncread(path3d[1],"Y")
nm=[]
# establish tracers name array
    for i in 1:39
        a = 60+i
        nm=push!(nm,string(a))
    end
    push!(nm,"0a")
    push!(nm,"0b")
    push!(nm,"0c")
    push!(nm,"0d")
    push!(nm,"0e")
    push!(nm,"0f")
    push!(nm,"0g")
    push!(nm,"0h")
    push!(nm,"0i")
    push!(nm,"0j")
    push!(nm,"0k")
for n in 1:7
index_lat = findall(x->x==la[n], lat)
tm=zeros(120)
p1=zeros(120,10)
p_total=zeros(5,10,120)
for i in 1:120
t = ncread(path3d[i],"T")
tm[i] = t[1]/3600.0/24.0-1442.0
for j in 1:5
for k in 1:10
    nmt=nm[10*(j-1)+k]
    p_temp = ncread(path3d[i],"TRAC"*nmt)
    p_total[j,k,i]=sum(p_temp[289:296,index_lat[1],1,1])
end
end
end
p1=sum(p_total,dims=1)
p2=zeros(10,120)
p2=transpose(p1[1,:,:])
pplot=plot(tm,p2,title="Biomass variation across a year of coccolithophores with different thermal types \n at "*string(la[n])*"N",
titlefontsize = 9,
label = "",
color = [:firebrick :red1 :orange :yellow1 :lime :green :darkslategray1 :skyblue :royalblue :blue2],
#color = [:blue2 :royalblue :skyblue :darkslategray1 :green :lime :yellow1 :orange :red1 :firebrick],
lw=2,xlabel="Days",ylabel="Concentration (kg/kg)")
    
savefig(pplot,"/home/weixuan/Desktop/Jupyter notebook/biomass results/Biomass BATS variation/Biomass variation across a year of coccolithophores with different thermal types \n at "*string(la[n])*"N.png")
end
```
